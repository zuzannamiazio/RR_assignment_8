# Breaking Bad: report on viewers statistics

### Description
<i>Breaking Bad</i> is an American TV show, which tells a story about a chemistry teacher who got diagnosed with <b>cancer</b>: Walter White. He then proceeds to ask his ex-student: Jesse Pinkman to help him with producing and selling drugs. Walter White has two children on the show and a wife, therefore he wants his family to have a stable future, so he needed to earn a lot of money in a short time period. With him being a <b>chemistry teacher</b>, his experience leads him to produce high quality drugs and a high quantity.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('BB_data.csv')
```

```python
data.columns
```

```python
viewers_stats = data['U.S. viewers (millions)'].describe()
viewers_stats
```

```python
data['Original air date'] = pd.to_datetime(data['Original air date'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Original air date'], data['U.S. viewers (millions)'], marker='o', linestyle='-')
plt.title('Views Over Time for Breaking Bad')
plt.xlabel('Date')
plt.ylabel('U.S. Viewers (millions)')
plt.xticks(rotation=45)  
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
season_viewership = data.groupby('Season')['U.S. viewers (millions)'].sum()

plt.figure(figsize=(10, 6))
season_viewership.plot(kind='bar', color='skyblue')
plt.title('Season-to-Season Changes in Viewership')
plt.xlabel('Season')
plt.ylabel('Total U.S. Viewership (millions)')
plt.xticks(rotation=45) 
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Statistics summary

<b>The series has become more and more popular with each season.</b> 

Between Season 1 and Season 2 there has been a rise in viewership of: $8.37$ million, and between Season 4 and Season 5 the rise was: $44.78$  million viewers.
